# Contexto de aplicación



# Objetivo

Queremos predecir la demanda futura con base en los datos históricos de los registros de demanda.

y= demanda [$ \$ /MWh$]

x= fecha `MM/DD/YYYY` y hora `range(1,25,1)`

El dataset contiene la demanda total diaria, que consiste en ofertas de demanda fijas y sensibles al precio, más ofertas decrecientes e incrementales. En función, de la hora según varios días.

El dataset contiene variables exógenas, como temperatura y componentes del precio que afectan la demanda, estas variables podrían ser consideradas para mejorar el pronóstico de demanda energética.











# Dataset

El dataset es tomado de kaggle, con la demanda de energía ISO *New England Control Area (ISO-NE CA) and the eight load zones*, la cual es una área de Estados Unidos, que consume mayor energía que el todo Colombia.

Tamaño en disco: 19.1 MB

Todos los precios están dados en $/MWh.

A continuación son tomadasb textualmente las descripciones de los encabezados del dateset.



*   Date is the date in MM/DD/YYYY format.
*   Hour is the hour ending value with Hour 1 equal to the hour ending at 1:00 am.
*   DA_LMP is the day ahead locational marginal price.
*  DA_EC is the energy component of the day ahead price.
* DA_CC is the congestion component of the day ahead price.
* DA_MLC is the marginal loss component of the day ahead price.
* RT_LMP is the real time locational marginal price.
* RT_EC is the energy component of the real time price.
* RT_CC is the congestion component of the real time price.
* RT_MLC is the marginal loss component of the real time price.
* DryBulb is the dry bulb temperature in degrees Fahrenheit for the corresponding weather station.
* DewPt is dew point temperature in degrees Fahrenheit for the corresponding weather station.
